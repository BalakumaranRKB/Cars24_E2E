```python
def list_cleaner3(list_ij):
    list_copy = copy.deepcopy(list_ij)
    for i in range(0,len(list_copy)):
        if len(list_copy[i]) == 7:
            del list_copy[i][0]
        elif len(list_copy[i]) == 8:
            del list_copy[i][0:2]
        else:
            continue
    return list_copy
```

```python
def list_cleaner4(list_ij):
    list_copy = copy.deepcopy(list_ij)
    for i in range(0,623):
        if len(list_copy[i]) != 7:
            del list_copy[i]
        else:
            continue
    return list_copy
```

```python
def sanity_check_for_list(new_list1):
    count = 0
    for i in range(0,len(new_list1)):
        if len(new_list1[i]) != 7:
            print(len(new_list1[i]),i)
            count = count + 1
        else:
            continue
    print(count)
```

```python
def text_conversion_in_list(new_list1):
    test_list =  copy.deepcopy(new_list1)
    for i in range(0,len(test_list)):
        for j in range(0,len(test_list[i])):
            if (j == 0):
                y = test_list[i][0].split()[0]
                b = test_list[i][0].split()[1] + '-' + test_list[i][0].split()[2]
            elif (j == 3):
                k = test_list[i][3].split()[0]
                x = k.split()[0]
                x = int(x.replace(',',''))
                
            elif (j == 6):
                p = test_list[i][6]
                c = int(p.replace('₹','').replace(',',''))
            else:
                continue  
        test_list[i].extend((b, x, c,y))
    return test_list
```
